```python
import numpy as np
import re
from nltk.tokenize import word_tokenize
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
skl=set(ENGLISH_STOP_WORDS)
import nltk
from nltk.corpus import stopwords
nltk_stopswords=set(stopwords.words('english'))
```

```python
import nltk
nltk.download('stopwords')
```

```python
data['review'] = data['review'].str.lower()
data['review'] = data['review'].str.replace(r'@\S+|https?:\S+|[^a-zA-Z0-9\s]', '', regex=True)
```

```python
data
```

```python
combined_stopwords =nltk_stopswords.union(skl)
```

```python
#text normalization: stemming or lemmotization(prefer)
from nltk.stem import WordNetLemmatizer
lemmatizer =WordNetLemmatizer()
```

```python
import nltk
nltk.download('wordnet')
```

```python
# Initialize the lemmatizer
lemmatizer = WordNetLemmatizer()

# Create an empty list to store the cleaned tokens
cleaned_text = []

# Assuming 'Cleaned' is a column in your DataFrame
for index, row in data.iterrows():
    text = row['review']
    tokens = text.split()
    cleaned_tokens = [lemmatizer.lemmatize(token) for token in tokens if token not in combined_stopwords]
    cleaned_text.append(' '.join(cleaned_tokens))
```

```python

# Assign the cleaned text back to the DataFrame
data['Cleaned'] = cleaned_text
```

```python
def polarity(df):
    sid_obj=SentimentIntensityAnalyzer()
    sentiment_scores = sid_obj.polarity_scores(df)
    return sentiment_scores
```

```python
data['sentiment_scores']= data['Cleaned'].apply(polarity)
```

```python
def sentiment(label):
  if label['neg'] ==1.0:
    return "Negative"
  elif label['neu'] ==1.0:
    return "Neutral"
  elif label['pos'] ==1.0:
    return "positive"
  elif label['compound'] >0:
    return "positive"
  elif label['compound'] <0:
    return "Negative"
```

```python
data['Final']= data['sentiment_scores'].apply(sentiment)
```

```python
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame
counts = data['Final'].value_counts()
categories = counts.index

plt.pie(counts, labels=categories, autopct='%1.1f%%')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Sentiment')

plt.show()
```
